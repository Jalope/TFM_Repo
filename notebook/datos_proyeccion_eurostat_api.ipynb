```python
SEX_MAP = {
    'Females': 'F',
    'Males':   'M',
    'Total':   'T'
}

# escenario proyecciones → modelo
SCENARIO_MAP = {
    'Baseline projections':                 'BSL',
    'Sensitivity test: lower fertility':    'LFRT',
    'Sensitivity test: lower mortality':    'LMRT',
    'Sensitivity test: higher migration':   'HMIGR',
    'Sensitivity test: lower migration':    'LMIGR',
    'Sensitivity test: no migration':       'NMIGR'
}
```

```python
# --------------- FUNCIÓN AUXILIAR -----------------
def clean_eurostat(df: pd.DataFrame, observed=True) -> pd.DataFrame:
    """Normaliza columnas al esquema común."""
    # a) Renombrar
    df = df.rename(columns={
        'TIME_PERIOD': 'year',
        'OBS_VALUE':   'population',
        'geo':         'cod_nuts',
        'age':         'age',
        'sex':         'sex'
    })

    # b) Mapear sexo
    df['sex'] = df['sex'].map(SEX_MAP)

    # c) Escenario
    if observed:
        df['scenario'] = 'OBS'
    else:
        df['scenario'] = df['projection'].map(SCENARIO_MAP)
        df = df.drop(columns='projection')

    # d) year y population numéricos
    df['year']       = df['year'].astype(int)
    df['population'] = pd.to_numeric(df['population'], errors='coerce')

    # e) nuts_level
    def _nuts_level(code):
        return 0 if (code == 'EU27_2020' or len(code) == 2) else 2
    df['nuts_level'] = df['cod_nuts'].apply(_nuts_level)

    # f) Orden final y tipos
    cols = ['year','age','sex','scenario','cod_nuts','nuts_level','population']
    df = df[cols].copy()
    df['age']  = df['age'].astype('category')
    df['sex']  = df['sex'].astype('category')
    df['scenario'] = df['scenario'].astype('category')
    return df
```

```python
# Normalizar históricos 1960-2024
df_hist_norm = clean_eurostat(df_hasta24, observed=True)

# Normalizar proyecciones 2022-2100
df_proj_norm = clean_eurostat(df_proyecciones, observed=False)
```
